```python
import pandas as pd
order_table = pd.read_csv('orders_table.csv')
order_product = pd.read_csv('order_product.csv')
product_table = pd.read_csv('product_table.csv')
ab_test = pd.read_csv('Samples AB.csv')
```

```python
# обʼєднаємо всі таблиці в один датафрейм

ab_testing = pd.merge(ab_test, order_table, on='customer_id', how='left')
ab_testing = pd.merge(ab_testing, order_product, on='order_id', how='left')
ab_testing = pd.merge(ab_testing, product_table, on='product_id', how='left')

# перевіряємо результат обʼєднання
ab_testing
```

```python
ab_testing['revenue']= ab_testing['quantity']*ab_testing['price']
ab_testing['test_orders']=ab_testing['order_date'].str [:7]
ab_testing=ab_testing[ab_testing['test_date']== '2023-07-01']
ab_testing=ab_testing[ab_testing['test_orders']== '2023-07']
ab_testing
```

```python
ab_testing1=ab_testing[ab_testing['sample_group']==1]
ab_testing1
```

```python
test_sample_1 = ab_testing1.groupby('customer_id')['revenue'].mean()
test_sample_1_filled = test_sample_1.fillna(0)
test_sample_1_filled
```

```python
ab_testing2=ab_testing[ab_testing['sample_group']==2]
ab_testing2
```

```python
test_sample_2 = ab_testing2.groupby('customer_id')['revenue'].mean()
# Заміна NaN на 0
test_sample_2_filled = test_sample_2.fillna(0)
test_sample_2_filled
```

```python
# Вирівнювання розмірів груп
min_length = min(len(test_sample_1_filled), len(test_sample_2_filled))
test_data1 = test_sample_1_filled.iloc[:min_length]
test_data2 = test_sample_2_filled.iloc[:min_length]
```

```python
#Генеруємо вибірки з заданими параметрами
sample_1 = test_data1
sample_2 = test_data2
```

```python
import statsmodels
from statsmodels.stats.weightstats import ztest 
z_st, p_val = ztest(sample_2, sample_1)
print(f'z-статистика склав {round(z_st,2)}')
print(f'Маємо P-value {round(p_val,2)}')
```

```python
diff = test_data1.values - test_data2.values

from statsmodels.stats.weightstats import DescrStatsW
# Обчислення довірчого інтервалу
if len(diff) > 1:
    conf_int = DescrStatsW(diff).tconfint_mean(alpha=0.05)
    print('95%% confidence interval=[%.4f,%.4f]' % (conf_int[0], conf_int[1]))
else:
    print("Недостатньо даних для обчислення довірчого інтервалу.")
```
